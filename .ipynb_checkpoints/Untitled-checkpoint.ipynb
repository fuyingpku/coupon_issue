```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the library and get the overview first of all
df = pd.read_csv(r"\Users\klxd537\desktop\coupons.csv")
df.head(10)
```

```python
#find all the features
df.columns
```

```python
#get the numbers of rows and features 
df.shape
```

```python
#data info especially Dtype for future analysis and no need to remove null in this case
df.info()
```

```python
#define what we want to use to study the relationship
features = ['destination', 'passanger', 'weather', 'temperature', 'time', 'coupon',
       'expiration', 'gender', 'age', 'maritalStatus', 'has_children',
       'education', 'occupation', 'income', 'car', 'Bar', 'CoffeeHouse',
       'CarryAway', 'RestaurantLessThan20', 'Restaurant20To50',
       'toCoupon_GEQ5min', 'toCoupon_GEQ15min', 'toCoupon_GEQ25min',
       'direction_same', 'direction_opp']
```

```python
# get full list of the dataframe and make the list of features for us to make a panel of figures
list(enumerate(features))
```

```python
#font was too small to we have to zoom in, since too much info, we have to do this style to data screening
#get overview of all the features and with color coding organge for yes to go to take the coupon 
#please zoom in and read, more anaylysis is followed, this is for overview first to find the best hit
plt.figure(figsize=(100,100))
for i in enumerate(features):
    plt.subplot(5,5, i[0]+1)
    sns.countplot(x=i[1], hue="Y", data = df)
    plt.xlabel(i[1],fontsize=45) 
    plt.xticks(rotation=45)
    
```

```python
# just check colours 
#sns.color_palette()
```

```python
# plot a histogram to show temperature column in data with legend and x-axis label tilted at 60 degrees with title and x-axis and y-axis labels
sns.histplot(df, x='weather', color='blue', kde=True, legend=True, linewidth=1.5)
plt.title('How weather affect taking a coupon')
plt.xlabel('Weather')
plt.ylabel('Count')
plt.show()
```

```python
#previous ploting does not have the overview of all the coupon types, now it is good to calculate the coupon accpetance rate
coupon_accept = df.query('Y==1')['Y'].count()
coupon_total = df['Y'].count();
coupon_acccpetance_rate = coupon_accept/coupon_total*100
coupon_acccpetance_rate
#it is interesting to see only slightly more than half people will take free coupon, why? the future work of the questinaire should have this focus
```

```python
#the previous panel did not get a clear view of different coupon types, however, this is just to show there is likely no major preference of delivery certain types of coupon in this dataset
sns.histplot(df, x='coupon', color='blue',legend=True)
plt.xticks(rotation=60)
plt.show()
```
